# Plots for action values

```python
from suppl import save_fig, number_to_letter
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman"]
```

```python
file_names = [
    './history/action_stack_agent_gcnn_2023-05-07_16-16_test_5_intersection.csv',
    './history/node_stack_agent_gcnn_2023-05-07_16-16_test_5_intersection.csv'
]

timestamp = '_'.join(file_names[0].split('/')[2].split('.')[0].split('_')[-2:])

dfs = [pd.read_csv(x, sep='\t', header=None) for x in file_names]

dfs = [df.replace(-np.inf, np.nan) for df in dfs]
```

```python
plt.figure(figsize=(6,6))
ax = sns.boxplot(dfs[1])
ax.set_xticklabels([number_to_letter(x) for x in range(10)])
plt.ylabel('$V(s, node)$', size=13)
plt.xlabel('Node', size=13)
save_fig(f'node_rewards_plot_gcnn_2023-05-07_16-16_test_5_intersection')
plt.show()
```

```python
plt.figure(figsize=(6,6))
ax = sns.boxplot(dfs[0])
ax.set_xticklabels(['add_lane', 'remove_lane', 'add_righthand', 'add_roundabout', 'add_trafficlight'])
plt.ylabel('$V(s, action)$', size=13)
plt.xlabel('Action', size=13)
save_fig(f'action_rewards_plot_gcnn_2023-05-07_16-16_test_5_intersection')
plt.show()
```
